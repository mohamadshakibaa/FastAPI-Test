```python
from fastapi import FastAPI, Query
from typing import Optional
from pydantic import BaseModel

app = FastAPI()

@app.get("/items/1")
async def read_item(q: Optional[str] = Query(None, min_length=3, max_length=100)):
    result = {"items": [{"name": "bar"}, {"name": "foo"}]}
    if q:
        result.update({"q": q})
    return result
```
